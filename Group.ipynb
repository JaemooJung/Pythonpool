```
import random
import string

while True:
  people_num = input('총 인원수를 입력해주세요 : ')
  m_num = input('조당 인원을 입력해 주세요 : ')

  if (people_num.isdigit() != True) or (m_num.isdigit() != True): # input 값이 숫자인지 확인하기
    print('잘못된 입력입니다. 숫자를 입력해 주세요.')
    continue
  elif int(people_num)//int(m_num) > 26**2 + 26: #최대 범위 지정
    print('지정된 범위를 초과한 값입니다.')
  else:
    break

people_num = int(people_num)
m_num = int(m_num)

total = list(range(1,people_num + 1))

#랜덤으로 조 편성
group = []
total_group = [[]]*(people_num//m_num)
for j in range(people_num//m_num):
  for i in range(m_num):
    n = random.choices(total)[0] #랜덤값 뽑기
    group.append(n) #랜덤으로 뽑힌 값 그룹에 추가
    total.remove(n) #이미 뽑힌 값 total에서 지우기
  total_group[j] = group #total_group에 j번 그룹 할당
  group = [] #group 초기화

# A 부터 ZZ까지 알파벳 리스트 만들기 . string 모듈 사용
alphabet = list(string.ascii_uppercase)
aa_to_zz = []
for i in alphabet:     
  for j in alphabet:
    aa_to_zz.extend([i+j]) # i번째 알파벳 + j번째 알파벳을 리스트에 추가
alphabet.extend(aa_to_zz) #기존 A-Z 리스트에 AA-ZZ추가

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('<<조원 선정 결과>>')
for i in range(people_num//m_num): #조 이름과 리스트 묶어서 출력
  total_group[i] = sorted(total_group[i])
  print(f'{alphabet[i]}조:',total_group[i])

if len(total) != 0: #조에 포함되지 못한사람 출력
  sorted(total)
  print('조가 없는 사람:',total)
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
```
